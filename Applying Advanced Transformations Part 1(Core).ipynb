```python
cat_cols = df
for col in cat_cols:
  display(df[col].value_counts())
```

```python
df_info['Hero|Publisher'].head(2)
```

```python
## adding expand=True
df_info['Hero|Publisher'].str.split('|',expand=True)
```

```python
## save the 2 new columns into the dataframe
df_info[['Hero','Publisher']] = df_info['Hero|Publisher'].str.split('|',expand=True)
df_info.head(2)
```

```python
# Make a list of all characters to replace
to_replace = ["{","}", "'", "'", "Height:", "Weight:","cm", "kg"]
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    df_info['Measurements'] = df_info['Measurements'].str.replace(char,'',regex=False)
    
df_info['Measurements'].head()
```

```python
df_info[['Height (cm)','Weight (kg)']] = df_info['Measurements'].str.split(',',expand=True)
df_info.head(2)
```

```python
df_info = df_info.drop(columns=['Hero|Publisher', 'Measurements'])
df_info.head(2)
```

```python
df_info["Height (cm)"] = df_info[["Height (cm)"]].astype(float)
```

```python
df_info["Height (cm)"].head()
```

```python
df_info["Weight (kg)"].astype(float)
df_info["Weight (kg)"].head()
```

```python
df_info.info()
```

```python
df = pd.concat((df, df_info), axis = 1)
df.head(2)
```

```python
df["Hero Names"] = df["hero_names"]
df = df.drop(columns=["hero_names", "Hero"])
df.head(2)
```

```python
## showing the lists are really strings
df.loc[0,'Powers']
```

```python
## adding expand=True
# Create a new column where they were converted back to lists
df["powers_split"] = df['Powers'].str.split(',')
df["powers_split"].value_counts()
```

```python
exploded = df.explode('powers_split')
```

```python
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
  df[col] = df['Powers'].str.contains(col)
df.head()
```

```python
# drop transactions columns
df = df.drop(columns=['Powers','powers_split'])
## save data for next lesson
df.to_csv('advanced_ps_data_pt1.csv', index=False)
```
